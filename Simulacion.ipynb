```python
#PASO 0. INSTALACION DE LIBRERIAS
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
#PASO 1. LEER LA BASE DE DATOS
ERDB = pd.read_csv('EarlyRetirementPrediction (1).csv')
ERDB.head()
```

```python
#DEFINIR CONJUNTO DE ENTRENAMIENTO Y CONJUNTO DE PRUEBA

train, test = train_test_split(ERDB, test_size=0.2)
```

```python
print(test)
```

```python
train_x=train[["Gender","Desease","Education Level","Marital Status",
           "Monthly Income","Employee/Employer","Dependants",
           "Unemployment Rate","Stock Market","Credit Score","Government Bonds Return"]]
train_y=train[["Retire Before 65 Years Old"]]
test_x=test[["Gender","Desease","Education Level","Marital Status",
           "Monthly Income","Employee/Employer","Dependants",
           "Unemployment Rate","Stock Market","Credit Score","Government Bonds Return"]]
test_y=test[["Retire Before 65 Years Old"]]
```

```python
#DEFINE SEQUENTIAL MODEL
model = Sequential()
model.add(Dense(4, input_dim=11, activation='sigmoid'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='mean_squared_error', optimizer='SGD', metrics=['accuracy'])
```

```python
model.fit(train_x, train_y, epochs=150, batch_size=10)
```

```python
# evaluate the keras model
print("Evaluate on test data")
results = model.evaluate(test_x, test_y)
print("test loss, test acc:", results)
```

```python
# Generate predictions (probabilities -- the output of the last layer)
# on new data using `predict`
print("Generate predictions for 3 samples")
predictions = model.predict(test_x)
print("predictions shape:", predictions)
```

```python
PredictionsDF=pd.DataFrame(predictions)
PredictionsDF.columns = ["Probabilities"]
PredictionsDF.head(20)
```

```python
PredictionsDF["Umbral"]=0
PredictionsDF.loc[PredictionsDF["Probabilities"] >0.5, "Umbral"] = 1
PredictionsDF.head(20)
```

```python
test_y=pd.DataFrame(test_y)
test_y.head()
```

```python
PredictionsDF.reset_index(drop=True,inplace=True)
test_y.reset_index(drop=True,inplace=True)
print(test_y)
```

```python
data=pd.concat([PredictionsDF,test_y],axis=1)
data=pd.DataFrame(data)
data.head()
```

```python
data["PositivePositive"]=0
data["Sum1"]=data["Umbral"]+data["Retire Before 65 Years Old"]
data.loc[data["Sum1"] ==2, "PositivePositive"] = 1
data.head()
```

```python
TPR=data["PositivePositive"].sum()/data["Retire Before 65 Years Old"].sum()
TPR
```

```python
data["NegativeNegative"]=0
data.loc[data["Sum1"] ==0, "NegativeNegative"] = 1
data.head()
```

```python
TNR=data["NegativeNegative"].sum()/(300-data["Retire Before 65 Years Old"].sum())
TNR
```

```python
BA=(TPR+TNR)/2
BA
```

```python
ACC=(data["NegativeNegative"].sum()+data["PositivePositive"].sum())/300
ACC
```
